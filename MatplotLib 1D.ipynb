```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x=[1,2,3,4,5,6,7]
y=[50,51,52,48,47,49,46]
```

```python
plt.xlabel("Days")
plt.ylabel("Temperature")
plt.title("Weather")
# plt.plot(x,y,color = "green",linewidth=5, linestyle='dashdot')
plt.plot(x,y,"gd--")
```

```python
plt.plot(x,y,color = "#888888",marker = "D", linestyle='',markersize = 15)
```
