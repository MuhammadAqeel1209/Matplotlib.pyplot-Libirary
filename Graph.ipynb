# bar Chart

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

```python
company=['GOOGL','AMZN','MSFT','FB']
revenue=[90,136,89,27]
```

```python
xpos = np.arange(len(company))
xpos
```

```python
plt.bar(xpos,revenue, label="Revenue")

plt.xticks(xpos,company)
plt.ylabel("Revenue(Bln)")
plt.title('US Technology Stocks')
plt.legend()
```

```python
profit=[40,2,34,12]
```

```python
plt.bar(xpos,revenue, label="Revenue")
plt.bar(xpos,profit, label="Profit")


plt.xticks(xpos,company)
plt.ylabel("Revenue(Bln)")
plt.title('US Technology Stocks')
plt.legend()
```

```python
plt.bar(xpos-0.2,revenue,width=0.4,label="Revenue")
plt.bar(xpos+0.3,profit,width=0.4,label="Profit")


plt.xticks(xpos,company)
plt.ylabel("Revenue(Bln)")
plt.title('US Technology Stocks')
plt.legend()
```

```python
plt.barh(xpos-0.2,revenue, label="Revenue")
plt.barh(xpos+0.2,profit, label="Profit")


plt.yticks(xpos,company)
plt.xlabel("Revenue(Bln)")
plt.title('US Technology Stocks')
plt.legend()
```

# Histogram 

```python
blood_sugar = [113, 85, 90, 150, 149, 88, 93, 115, 135, 80, 77, 82, 129]
plt.hist(blood_sugar, rwidth=0.8)
```

```python
plt.hist(blood_sugar,rwidth=0.5,bins=4,color="g",histtype="step")
```

```python
plt.xlabel("Sugar Level")
plt.ylabel("Number Of Patients")
plt.title("Blood Sugar Chart")

blood_sugar_men = [113, 85, 90, 150, 149, 88, 93, 115, 135, 80, 77, 82, 129]
blood_sugar_women = [67, 98, 89, 120, 133, 150, 84, 69, 89, 79, 120, 112, 100]

plt.hist([blood_sugar_men,blood_sugar_women], bins=[80,100,125,150], rwidth=0.95, color=['green','orange'],label=['men','women'])
plt.legend()
```

# Pie Chart

```python
exp_vals = [1400,600,300,410,250]
exp_labels = ["Home Rent","Food","Phone/Internet Bill","Car ","Other Utilities"]
plt.axis("equal")
plt.pie(exp_vals,labels=exp_labels,radius=1.5 ,shadow=True, autopct='%0.0f%%',explode=[0,0,0,0,0])
plt.show()
```

```python
exp_vals = [1400,600,300,410,250]
exp_labels = ["Home Rent","Food","Phone/Internet Bill","Car ","Other Utilities"]
plt.axis("equal")
plt.pie(exp_vals,labels=exp_labels,radius=1.5 ,shadow=True, autopct='%0.0f%%',explode=[0,0,0,0.1,0],startangle=90)
plt.show()
plt.savefig("pieChart.png")
```
